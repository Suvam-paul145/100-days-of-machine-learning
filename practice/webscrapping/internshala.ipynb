```python
Internship_domain=[]
Company=[]
Location=[]
Stipend=[]
about_job=[]
for page in range(1,21):
    web = requests.get('https://internshala.com/internships/engineering-internship/page-{}'.format(page)).text
    soup = BeautifulSoup(web, 'lxml')
    company = soup.find_all('div',class_='internship_meta duration_meta')
    for i in company:
        try:
            Internship_domain.append(i.find('a', class_='job-title-href').text)
        except:
            Internship_domain.append(np.nan)

        try:
            Company.append(i.find('p', class_='company-name').text)
        except:
            Company.append(np.nan)

        try:
            Location.append(i.find('span>a', class_='row-1-item locations').text)
        except:
            Location.append('Work From Home')

        try:
            Stipend.append(i.find('span', class_='stipend').text)
        except:
            Stipend.append(np.nan)

        try:
            about_job.append(i.find('div', class_='text').text)
        except:
            about_job.append(np.nan)


df = pd.DataFrame(
    {
        'Internship_domain' : Internship_domain,
        'Company_name' : Company,
        'Location' : Location,
        'Stipend' : Stipend,
        'job_description': about_job
    }
)
                      
    # print(Internship_domain)
```
