```python
json_data = response.json()
```

```python
# Get the JSON data from response
weather_data = response.json()

# Examine the structure of the data
print("Keys in the response:", weather_data.keys())

# Create DataFrame from the 'list' key which contains weather forecasts
df = pd.json_normalize(weather_data['list'])

# Display the first few rows and columns of the DataFrame
print("\nDataFrame columns:", df.columns.tolist())
display(df.head())

# Create a more focused DataFrame with key weather information
weather_df = pd.DataFrame({
    'datetime': pd.to_datetime(df['dt_txt']),
    'temperature': df['main.temp'],
    'feels_like': df['main.feels_like'],
    'humidity': df['main.humidity'],
    'weather_description': df['weather'].apply(lambda x: x[0]['description']),
    'wind_speed': df['wind.speed']
})

```
