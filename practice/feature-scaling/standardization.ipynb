```python
dataset = df.iloc[:,2:]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(dataset.drop('Purchased', axis=1),dataset['Purchased'],test_size=0.3, random_state=42)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()

Scaler.fit(x_train)

x_train_scaled = Scaler.transform(x_train)
x_test_scaled = Scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
x_train_df = pd.DataFrame(x_train_scaled,columns=x_train.columns)

y_test_df = pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize=(12,5))
# fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))
ax1.scatter(x_train['Age'], x_train['EstimatedSalary'], color ='blue')
ax1.set_title("Before scalinh")
ax2.scatter(x_train_df['Age'], x_train_df['EstimatedSalary'], color ='red')
ax2.set_title("After scalinh")
```
